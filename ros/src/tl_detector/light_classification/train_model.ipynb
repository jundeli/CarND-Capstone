```python
import numpy as np
import cv2
import os
from glob import glob
```

```python
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Flatten, Lambda,  Dense, Dropout, Conv2D, MaxPooling2D, Activation
from keras.optimizers import Adam
```

```python
categories = [ [1,0,0] , [0,1,0], [0,0,1]]  # Red, Yellow, Green
```

```python
features = []
labels = []
paths = glob(os.path.join('data', '*.png'))
for path in paths:
    cat_idx = int(path[-5])
    if 0 <= cat_idx <= 2:
        img = cv2.imread(path)
        features.append(img)
        labels.append( categories[cat_idx] )
features = np.array(features)
labels = np.array(labels)
```

```python
features.shape
```

```python
labels.shape
```

```python
model = Sequential()
# Normalization
model.add( Lambda(lambda x: x/255.0 - 0.5, input_shape=(64,32,3)))
# LeNet - based architecture
model.add( Dropout(.2))
model.add( Conv2D( 6, (5, 5), padding='same', activation='relu') )
model.add( MaxPooling2D() )
model.add( Dropout(.2))
model.add( Conv2D( 6, (5, 5), padding='same',  activation='relu') )
model.add( MaxPooling2D() )
model.add( Flatten())
model.add( Dropout(.2))
model.add( Dense(120) )
model.add( Dropout(.2))
model.add( Dense(60, activation='relu') )
#softmax output
model.add(Dense(3))
model.add(Activation("softmax"))

model.summary()
```

```python
optimizer = Adam(lr=0.0001)
model.compile(loss='categorical_crossentropy', optimizer=optimizer)
model.fit( features, labels, epochs=1000, validation_split=0.3, shuffle=True )
```

```python
model.save(os.path.join('models', 'lenet27.h5'))
```
